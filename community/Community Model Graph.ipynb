```python
from twitterSim_comm import *
from random import random
```

```python
def graph(G):
    import matplotlib.pyplot as plt
    pos = nx.spring_layout(G)
    edges = [ ((comm, otherComm) for otherComm in G.neighbors(comm)) for comm in G.nodes() ]
    color_map = [ random()*1000 for comm in G.nodes() ]
    nx.draw_networkx_nodes(G, pos, cmap=plt.get_cmap('jet'), node_size = 500, node_color=color_map)
    nx.draw_networkx_edges(G, pos, alpha=0.5, arrows=False)
    labels = { comm: comm.id for comm in G.nodes() }
    nx.draw_networkx_labels(G, pos, labels)
    plt.title("Community Model Graph")
    plt.show()
```

```python
G = runSim()
```

```python
graph(G)
```
